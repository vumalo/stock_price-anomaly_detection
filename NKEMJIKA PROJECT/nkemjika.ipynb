# Machine learning libraries needed

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import plot_confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

# Machine learning function

```python
def ml(mf, clt):
    x=mf['Suspected Organism']
    y=mf.drop('Suspected Organism',axis=1)
    X_train, X_test, y_train, y_test= train_test_split(y, x, test_size=0.25, random_state=42)
    clt.fit(X_train, y_train)
    accuracy=metrics.accuracy_score(y_test, clt.predict(X_test))
    return accuracy
    
    
```
